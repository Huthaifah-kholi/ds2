```python
# trying https://stackoverflow.com/questions/44487356/tweepy-search-giving-more-result-than-count-parameter
num_tweets=120
query = '"data journalism"' # this passes "tesla stock" as an exact phrase to search for

num = 0
# create list to append tweets to
tweets = []
result = tweepy.Cursor(api.search,q=query + " -filter:retweets",lang="en", show_user = True).items(num_tweets)
for tweet in result:
    tweets.append(tweet)
    num +=1
    print(num, tweet.user.name, "said on", tweet.created_at) #, "/", tweet.created_at.month, "/",
          #tweet.created_at.year, "at", tweet.created_at.timestamp)
    print(tweet.text, "\n---------------------------------")
if num < 1:
    print("No tweets found so far for current search criteria")

# write to file
# convert 'tweets' list to pandas.DataFrame
df = pd.DataFrame(vars(tweets[i]) for i in range(len(tweets)))
# define file path (string) to save csv file to
FILE_PATH = "tweets.csv"

# define attributes you want
tw_fields = [
'text', 'created_at', 'favorite_count',
'lang', 'retweet_count', 'source','retweeted',
'id'
]

# subset dataframe
df = df[tw_fields]
# use pandas to save dataframe to csv
df.to_csv(FILE_PATH)   
```

```python
# try also: https://stackoverflow.com/questions/22469713/managing-tweepy-api-search?rq=1
# mixed (default), recent, popular are result_type in api.search
query = 'python'
max_tweets = 1000
#searched_tweets = [status for status in tweepy.Cursor(api.search, q=query).items(max_tweets)] # memory intensive?
searched_tweets = []
last_id = -1
while len(searched_tweets) < max_tweets:
    count = max_tweets - len(searched_tweets)
    try:
        new_tweets = api.search(q=query, count=count, max_id=str(last_id - 1))
        if not new_tweets:
            break
        searched_tweets.extend(new_tweets)
        last_id = new_tweets[-1].id
    except tweepy.TweepError as e:
        # depending on TweepError.code, one may want to retry or wait
        # to keep things simple, we will give up on an error
        break
```
