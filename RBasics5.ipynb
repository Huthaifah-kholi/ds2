## R Basics 5: Graphics

```r
# iris is an internal dataset in R, stored as dataframe with 4 features and one target (5 columns)
str(iris)
```

```r
# for categorical data (factors), plot displays histogram. In iris, each species has 50 samples 
# if two categorical variables are plotted, an area plot is generated 
plot(iris$Species)
```

```r
# for numerical data, a scatter plot
# iris$Sepal.Length is the first column
plot(iris$Sepal.Length)
```

```r
# can specify 2-d plots 
plot(iris$Sepal.Length, iris$Sepal.Width)
```

```r
# histograms 
# let's plot setosa species 
setosa <- iris[iris$Species == "setosa",]
hist(setosa$Sepal.Length)
```

```r
# breaks controls the number of bins
hist(setosa$Sepal.Length, breaks = 20)
```

```r
# plot with options 
plot(setosa$Sepal.Length, setosa$Sepal.Width, 
    xlab = "Sepal Length",
    ylab = "Sepal Width",
    main = "Sepal Width vs. Sepal Length",
    #type = "o",
    col = "green")
```

```r
# par() sets and displays plot parameters 
par(col = "blue")
par()$col
```

```r
# control number of plots (number of rows, number of columns)
par(mfrow = c(2,2))
#par(mfcol = c(2,2))
#par(mfrow = c(1,1)) # default
# reset layout
#layout(1)
# define x and y as two vectors
x <- c(2.2, 3, 3.8, 4.5, 7, 8.5, 9.7, 11.5)
y <- c(4, 5.5, 6.5, 9, 11, 15.2, 18.3, 20.5)
plot(x, y,
xlab = "X",
ylab = "Y",
main = "Y vs. X",
type = "o",
col = "orange",
col.main = "darkblue",
cex.axis = 0.8,  # label size
lty = 3,  # dotted
pch = 0) # zero for square, defaul is 1 aka circle 

# all defaults, x^2 vs x
plot(x, x^2)
```

```r
lm_line <- lm(y ~ x)
plot(x, y,
xlab = "X",
ylab = "Y",
main = "Y vs. X",
type = "o",
col = "orange",
col.main = "darkblue",
cex.axis = 0.8,  # label size
lty = 3,  # dotted 
lwd = 4, # line width
pch = 0) # pch is symbol type, 0 is square
# add linear regression line
abline(coef(lm_line), lwd = 3)
```

```r
# you can find this and more at 
# http://www.sthda.com/english/wiki/r-plot-pch-symbols-the-different-point-shapes-available-in-r

#++++++++++++++++++++++++++++++++++++++++++++
#generate a plot of point shapes which R knows about.
#++++++++++++++++++++++++++++++++++++++++++++
generateRPointShapes <- function(){
  oldPar <- par()
  par(font=2, mar=c(0.5,0,0,0))
  y=rev(c(rep(1,6),rep(2,5), rep(3,5), rep(4,5), rep(5,5)))
  x=c(rep(1:5,5),6)
  plot(x, y, pch = 0:25, cex=1.5, ylim=c(1,5.5), xlim=c(1,6.5), 
       axes=FALSE, xlab="", ylab="", bg="blue")
  text(x, y, labels=0:25, pos=3)
  par(mar=oldPar$mar,font=oldPar$font )
}
generateRPointShapes()
```
